# **Statistics Part 2**

---
#1. What is hypothesis testing in statistics?
Hypothesis testing is a statistical method used to make inferences about a population based on sample data. It involves making an assumption (hypothesis) and testing whether it holds true using statistical techniques.

---

#2. What is the null hypothesis, and how does it differ from the alternative hypothesis?

  - Null Hypothesis ($H_0$): Assumes no effect or no difference.

  - Alternative Hypothesis ($H_1$ or $H_a$): Assumes there is an effect or a difference.
  
We test $H_0$ and try to find evidence to support $H_1$.

---

#3. What is the significance level in hypothesis testing, and why is it important?
The significance level ($\alpha$) is the threshold probability used to decide whether to reject the null hypothesis. Common values are $\alpha = 0.05$, $0.01$, or $0.10$. It controls the risk of a Type I error

---

#4. What does a P-value represent in hypothesis testing?
The P-value is the probability of obtaining a result equal to or more extreme than what was actually observed, assuming that the null hypothesis is true.

---

#5. How do you interpret the P-value in hypothesis testing?

  - If $p \leq \alpha$: Reject the null hypothesis (statistically significant).

  - If $p > \alpha$: Fail to reject the null hypothesis (not statistically significant).

---

#6. What are Type 1 and Type 2 errors in hypothesis testing?

  - Type I Error ($\alpha$): Rejecting a true null hypothesis.

  - Type II Error ($\beta$): Failing to reject a false null hypothesis.

---

#7. What is the difference between a one-tailed and a two-tailed test in hypothesis testing?

  - One-tailed test: Tests for the effect in one direction (greater than or less than).

  - Two-tailed test: Tests for the effect in both directions (not equal to).

---

#8. What is the Z-test, and when is it used in hypothesis testing?
A Z-test is used when the population variance is known and the sample size is large ($n \geq 30$). It tests whether a sample mean significantly differs from a known population mean.

---

#9. How do you calculate the Z-score, and what does it represent in hypothesis testing?
The Z-score is calculated as:

$$
Z = \frac{\bar{X} - \mu}{\frac{\sigma}{\sqrt{n}}}
$$

Where:

  - $\bar{X}$ = sample mean

  - $\mu$ = population mean

  - $\sigma$ = population standard deviation

  - $n$ = sample size

It represents how many standard deviations a data point is from the population mean.

---

#10. What is the T-distribution, and when should it be used instead of the normal distribution?
The T-distribution is used instead of the normal distribution when the sample size is small ($n < 30$) and/or the population standard deviation is unknown. It is wider and has heavier tails.

---

#11. What is the difference between a Z-test and a T-test?

  - Z-test: Used when population variance is known and sample size is large.

  - T-test: Used when population variance is unknown and/or sample size is small.

---

#12. What is the T-test, and how is it used in hypothesis testing?
The T-test compares sample means to determine if they are significantly different. Types include one-sample, two-sample (independent), and paired sample T-tests.

---

#13. What is the relationship between Z-test and T-test in hypothesis testing?
Both are used to compare means. The Z-test is a special case of the T-test used when sample sizes are large or population variance is known.

---

#14. What is a confidence interval, and how is it used to interpret statistical results?
A confidence interval (CI) is a range of values within which the true population parameter is likely to lie, with a certain level of confidence (e.g., 95%).

---

#15. What is the margin of error, and how does it affect the confidence interval?
The margin of error is the range added and subtracted from the sample statistic in a confidence interval. A larger margin of error results in a wider confidence interval.

---

#16. How is Bayes' Theorem used in statistics, and what is its significance?
Bayes’ Theorem is used to update the probability of a hypothesis based on new evidence:

$$
P(A|B) = \frac{P(B|A) \times P(A)}{P(B)}
$$

It's significant in decision-making under uncertainty.

---

#17. What is the Chi-square distribution, and when is it used?
The Chi-square distribution is used in tests of independence and goodness of fit for categorical data. It is positively skewed and only takes non-negative values.

---

#18. What is the Chi-square goodness of fit test, and how is it applied?
This test checks if the observed frequency distribution differs from the expected distribution. The test statistic is:

$$
\chi^2 = \sum \frac{(O - E)^2}{E}
$$

Where $O$ is observed frequency, and $E$ is expected frequency.

---

#19. What is the F-distribution, and when is it used in hypothesis testing?
The F-distribution is used to compare two variances. It’s mainly used in ANOVA and in the F-test for equality of variances.

---

#20. What is an ANOVA test, and what are its assumptions?
ANOVA (Analysis of Variance) compares means of three or more groups.
Assumptions:

  - Independence of observations

  - Normally distributed populations

  - Homogeneity of variances

---

#21. What are the different types of ANOVA tests?

  - One-Way ANOVA: One independent variable

  - Two-Way ANOVA: Two independent variables

  - MANOVA: Multiple dependent variables

  - Repeated Measures ANOVA: Same subjects used in all treatments

---

#22. What is the F-test, and how does it relate to hypothesis testing?
The F-test compares variances of two populations to see if they are significantly different. It's used in ANOVA and regression analysis to test model significance.

```python
# 15. Implement hypothesis testing using Z-statistics for a sample dataset.
import numpy as np
from scipy.stats import norm

# Sample data
sample = np.random.normal(50, 10, 30)
sample_mean = np.mean(sample)
population_mean = 52
population_std = 10
n = len(sample)

# Calculate Z-statistic
z = (sample_mean - population_mean) / (population_std / np.sqrt(n))

# Calculate p-value for two-tailed test
p_value = 2 * (1 - norm.cdf(abs(z)))

print(f"Sample mean: {sample_mean:.2f}")
print(f"Z-statistic: {z:.3f}")
print(f"P-value: {p_value:.4f}")
```

```python
# 16. Create a confidence interval for a dataset using Python and interpret the result.
import numpy as np
from scipy.stats import norm

data = np.random.normal(100, 15, 100)
confidence_level = 0.95
mean = np.mean(data)
std = np.std(data, ddof=1)
n = len(data)

z_critical = norm.ppf((1 + confidence_level) / 2)
margin_of_error = z_critical * (std / np.sqrt(n))

confidence_interval = (mean - margin_of_error, mean + margin_of_error)

print(f"Mean: {mean:.2f}")
print(f"Confidence Interval (95%): {confidence_interval}")
```

```python
# 1. Write a Python program to perform a Z-test for comparing a sample mean to a known population mean and interpret the results.
from scipy.stats import norm
import numpy as np

# Sample data
sample = np.array([101, 99, 98, 97, 103, 100, 102, 99])
sample_mean = np.mean(sample)
pop_mean = 100
pop_std = 2  # known population standard deviation
n = len(sample)

z = (sample_mean - pop_mean) / (pop_std / np.sqrt(n))
p = 2 * (1 - norm.cdf(abs(z)))

print(f"Z-score: {z:.3f}")
print(f"P-value: {p:.4f}")
```

```python
# 2. Simulate random data to perform hypothesis testing and calculate the corresponding P-value using Python.
from scipy.stats import ttest_1samp

data = np.random.normal(loc=5, scale=1, size=30)
t_stat, p_val = ttest_1samp(data, 5)
print(f"T-statistic: {t_stat:.3f}, P-value: {p_val:.4f}")
```

```python
# 3. Implement a one-sample Z-test using Python to compare the sample mean with the population mean.
def one_sample_z_test(sample, pop_mean, pop_std):
    n = len(sample)
    sample_mean = np.mean(sample)
    z = (sample_mean - pop_mean) / (pop_std / np.sqrt(n))
    p = 2 * (1 - norm.cdf(abs(z)))
    return z, p

sample = np.random.normal(loc=50, scale=5, size=30)
z, p = one_sample_z_test(sample, 50, 5)
print(f"Z-score: {z:.3f}, P-value: {p:.4f}")
```

```python
# 5. Create a Python function that calculates and visualizes Type 1 and Type 2 errors during hypothesis testing.
def visualize_errors(mu0, mu1, sigma, n, alpha):
    x = np.linspace(mu0 - 4*sigma, mu1 + 4*sigma, 1000)
    z_critical = norm.ppf(1 - alpha/2)
    threshold = mu0 + z_critical * sigma / np.sqrt(n)

    y0 = norm.pdf(x, mu0, sigma / np.sqrt(n))
    y1 = norm.pdf(x, mu1, sigma / np.sqrt(n))

    plt.plot(x, y0, label='H0 Distribution')
    plt.plot(x, y1, label='H1 Distribution')
    plt.axvline(threshold, color='black', linestyle='--', label='Critical value')
    plt.fill_between(x, y0, where=(x > threshold), color='red', alpha=0.3, label='Type I Error')
    plt.fill_between(x, y1, where=(x < threshold), color='blue', alpha=0.3, label='Type II Error')
    plt.legend()
    plt.title("Type I and Type II Errors")
    plt.show()

visualize_errors(mu0=100, mu1=102, sigma=10, n=30, alpha=0.05)
```

```python
# 9. Write a Python function to calculate the confidence interval for a sample mean and explain its significance.
from scipy.stats import norm

def confidence_interval(data, confidence=0.95):
    n = len(data)
    mean = np.mean(data)
    std_err = np.std(data, ddof=1) / np.sqrt(n)
    margin = norm.ppf((1 + confidence) / 2) * std_err
    return mean - margin, mean + margin

data = np.random.normal(50, 10, 100)
ci = confidence_interval(data)
print("Confidence Interval:", ci)
```

```python
# 10. Write a Python program to calculate the margin of error for a given confidence level using sample data.
def margin_of_error(data, confidence=0.95):
    n = len(data)
    std_err = np.std(data, ddof=1) / np.sqrt(n)
    return norm.ppf((1 + confidence) / 2) * std_err

sample_data = np.random.normal(70, 15, 50)
moe = margin_of_error(sample_data)
print("Margin of Error:", moe)
```

```python
# 11. Implement a Bayesian inference method using Bayes' Theorem in Python and explain the process.
def bayes_theorem(prior_A, likelihood_B_given_A, prior_B):
    return (likelihood_B_given_A * prior_A) / prior_B

# Example:
P_A = 0.01  # Disease prevalence
P_B_given_A = 0.9  # Test is positive if diseased
P_B = 0.05  # Overall probability of positive test

posterior = bayes_theorem(P_A, P_B_given_A, P_B)
print("Posterior Probability P(A|B):", posterior)
```
